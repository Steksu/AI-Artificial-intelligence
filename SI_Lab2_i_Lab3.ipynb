```python
# KROK 1 losowanie paramterow a i b
a = random.uniform(-10,10)
b = random.uniform(-10,10)
print('Parametry a i b maja wartosci',a,b)
```

```python


#y_pred - y


n = len(x)
learning_rate = 0.001

listaa = [a]
listab = [b]
listaMSE = []

for i in range(0,100000):
  listaa.append(a)
  listab.append(b)
  # KROK 2 obliczenie aktualnej warosci cisninenia przewidywanego
  y_predicted = list(map(lambda iks: a * iks + b, x))

  # KROK3 Obliczenie pochodnych dJ_da oraz dJ_db
  roznica_ypred_y_razy_x = []

  roznica_y_ypred_do_kwadratu = []


  for i in range(0, len(x)): #i przyjmie wartosci od 0 do 9
    roznica_ypred_y_razy_x.append( (y_predicted[i]-y[i])*x[i]) #append dodaje roznice ypred_i y_i
    roznica_y_ypred_do_kwadratu.append( ( y_predicted[i] - y[i] )**2 )

  mse = 1/n * sum(roznica_y_ypred_do_kwadratu)
  listaMSE.append(mse)

  dJ_da = 2/n * sum(roznica_ypred_y_razy_x)


  #dJ_db
  roznica_ypred_y = []

  for i in range(0, len(x)): #i przyjmie wartosci od 0 do 9
   roznica_ypred_y.append(y_predicted[i]-y[i]) #append dodaje roznice ypred_i y_i


  dJ_db = 2/n * sum(roznica_ypred_y)

  #KROK 4 atualizacja wartosci a i b
  a = a - learning_rate * dJ_da
  b = b - learning_rate * dJ_db





print('dJ_da= ',dJ_da, 'dJ_db= ', dJ_db, 'a= ',a, 'b= ',  b)
print('Zmiana a ', listaa[:100])
print('Zmiana b', listab[:100])
print('Zmiana MSE', listaMSE[:100])

```

```python
import csv #biblioteka do obslugi plikow CSV
import matplotlib.pyplot as plt


#pobranie pliku CSV
!wget bigbang.prz.edu.pl/si/birth_weight_vs_iq.csv


#wczytujemy plik CSV o podanej nazwie\
with open('birth_weight_vs_iq.csv') as csv_file:
    #ustawiamy znak rozdzielajacy wartosci w pliku
    #w tym przypadku to przecinek
    csv_reader = csv.reader(csv_file, delimiter=',')

    #tworzymy puste tablice do zapamietania
    #wagi urodzeniowej dziecka
    #to zbior wartosci x
    birth_weight=[]
    #wartosci IQ w wieku 5 lat
    #to zbior wartosci y
    iq_value=[]


    #odczytujemy plik CSV wiersz po wierszu
    for row in csv_reader:
      #w zmiennej row jest zapamietana zawartosc wiersza
      #w row[0] - zapamietana jest wartosc wagi (pierwsza kolumna)
      birth_weight.append( float(row[0]) )
      #w row[1] - zapamietany jest IQ (druga kolumna)
      iq_value.append( float(row[1]) )


#rysuhjemy wykres zeby sprawdzic czy import sie udal
plt.scatter(birth_weight, iq_value)

```

```python
print(len(birth_weight))
print(len(iq_value))
```

```python
plt.xlabel('Waga dziecka')
plt.ylabel('IQ dziecka')
plt.scatter(birth_weight, iq_value, color="blue")
plt.title(" IQ dziecka vs Waga dziecka ")
plt.legend()
plt.show()
```

Obliczenie optymalnej wartości parametru a z metody najmniejszych kwadratów

```python

x = list(map(lambda x: x / 1000, birth_weight))
y = list(map(lambda x: x / 100,  iq_value ))

#tworzymy pusta liste xy zeby zapamietamy iloczyny x1*y1, x2*y2 itd w liscie xy
xy = []

for i in range(0, len(x)):
  xy.append( x[i]*y[i] )

#wyswietlamy zawartosc listy xy
print(xy)

#sumujemy ilocznyny, aby policzyc 1 czlon licznika we wzorze na a
suma_xy = sum(xy)
print(suma_xy)
```

```python
xx = []

for i in range(0,len(x)):
  xx.append(x[i]**2)

print(x)

suma_xx = sum(xx)
```

```python
n = len(x)
a = (n * suma_xy - sum(x)*sum(y) ) / (n * suma_xx - sum(x)**2)
print(n)
print(a)
```

*Obliczenie* optymalnej wartości parametru b z metody najmniejszych kwadratów

```python
b = 1/n * (sum(y) - a*sum(x))
print(b)
```

```python
#w tej liscie zapamietamy przewidywane wartosci cisnienia od dawki leku
y_predicted = []


for i in range(0, len(x)): #i przyjmie wartosci od 0 do 9
  y_predicted.append( a*x[i]+b )


print(y_predicted)
```

```python
#dane prawdziwe
plt.scatter(x, y, color="black")
#dane przewidywane
plt.xlabel('Waga dziecka')
plt.ylabel('IQ dziecka')

#plt.scatter(x, y_predicted, label="predicted", color="red" )
plt.plot(x, y_predicted, color="red", label="prediction")

plt.title(" IQ dziecka vs Waga dziecka ")
plt.legend()
plt.show()
```

```python
from scipy import stats
from scipy.stats import chisquare
import statistics

#tworzymy liste do przechowywania kwadratow roznic
roznica_y_ypred_do_kwadratu = []


for i in range(0, len(x)): #i przyjmie wartosci od 0 do 9
  roznica_y_ypred_do_kwadratu.append( ( y_predicted[i] - y[i] )**2 ) #append dodaje roznice yi - ypred_i podniesiona do kwadratu


#wyswietlamy wartosc bledu mse
mse = 1/n * sum(roznica_y_ypred_do_kwadratu)
print("MSE", mse)
# R_Squared
variance_y = np.var(y)
R_squared = 1 - (mse/variance_y)
print("R^2", R_squared)
# Chi_Squared
chi_squared = np.sum(np.square(np.divide(np.subtract(y,y_predicted),y_predicted)))

print("Chi-squared:", chi_squared)

```

```python
# porownanie
from sklearn.metrics import mean_squared_error, r2_score

# Mean Squared Error (MSE)
mse_compare = mean_squared_error(y, y_predicted)
print("Mean Squared Error (MSE):", mse_compare)

# R^2
r_squared_compare = r2_score(y, y_predicted )
print("R-squared (R^2):", r_squared_compare)
```

Obliczyć pochodne dJ/da oraz dJ/db za pomocą pętli for, tak jak to robiono na poprzednich zajęciach. Możesz wykorzystać pętlę for do obliczania sumy iloczynów współrzędnych xi * yi.

```python
# KROK 1 losowanie paramterow a i b
a = random.uniform(-1,1)
b = random.uniform(-1,1)
print('Parametry a i b maja wartosci',a,b)
```

```python
n = len(x)
learning_rate = 0.001

listaa = [a]
listab = [b]
listaMSE = []

for i in range(0,10000):
  listaa.append(a)
  listab.append(b)
  # KROK 2 obliczenie aktualnej warosci cisninenia przewidywanego
  y_predicted = list(map(lambda iks: a * iks + b, x))

  # KROK3 Obliczenie pochodnych dJ_da oraz dJ_db
  roznica_ypred_y_razy_x = []

  roznica_y_ypred_do_kwadratu = []


  for i in range(0, len(x)): #i przyjmie wartosci od 0 do 9
    roznica_ypred_y_razy_x.append( (y_predicted[i]-y[i])*x[i]) #append dodaje roznice ypred_i y_i
    roznica_y_ypred_do_kwadratu.append( ( (y_predicted[i] - y[i]) )**2 )

  mse = 1/n * sum(roznica_y_ypred_do_kwadratu)
  listaMSE.append(mse)

  dJ_da = 2/n * sum(roznica_ypred_y_razy_x)


  #dJ_db
  roznica_ypred_y = []

  for i in range(0, len(x)): #i przyjmie wartosci od 0 do 9
   roznica_ypred_y.append(y_predicted[i]-y[i]) #append dodaje roznice ypred_i y_i


  dJ_db = 2/n * sum(roznica_ypred_y)

  #KROK 4 atualizacja wartosci a i b
  a = a - learning_rate * dJ_da
  b = b - learning_rate * dJ_db

print('dJ_da= ',dJ_da, 'dJ_db= ', dJ_db, 'a= ',a, 'b= ',  b)
print('Zmiana a ', listaa[:100])
print('Zmiana b', listab[:100])
print('Zmiana MSE', listaMSE[:100])
```
